```python
import requests
from bs4 import BeautifulSoup as bts

import re

import pymongo

from dotenv import load_dotenv
import os

import time
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
def get_friends(url):
    soup = getAndParseURL(url+"/friends/")
    friends_list=[]
    friends_ingame = soup.findAll('div', {'class': 'selectable friend_block_v2 persona in-game'})
    friends_online = soup.findAll('div', {'class': 'selectable friend_block_v2 persona online'})
    friends_offline = soup.findAll('div', {'class': 'selectable friend_block_v2 persona offline'})
    friends = friends_ingame + friends_online + friends_offline
    for friend in friends:
        friends_list.append(friend.a['href'])
    return friends_list
```

```python
tyler_friends = get_friends("https://steamcommunity.com/id/tylermcvicker/friends/")
len(tyler_friends)
```

```python
users_list = tyler_friends
for url in tyler_friends:
    print(url)
    users_list += get_friends(url)
```

```python
len(users_list)
```

```python
list_set = set(users_list)

users = (list(list_set))
```

```python
len(users)
```

```python
load_dotenv()
mongo_username = os.getenv("mongo_username")
mongo_password = os.getenv("mongo_password")

client = pymongo.MongoClient(f"mongodb+srv://{mongo_username}:{mongo_password}@cluster0.d4ojg.mongodb.net/myFirstDatabase?retryWrites=true&w=majority")
client.list_database_names()
```

```python
db = client.SteamDB
db.list_collection_names()
```

```python
user_collection = db.steam_users
user_games_collection = db.steam_user_games
user_reviews_collection = db.steam_user_reviews
```

```python
count=30000
for user in users[count:]:
    count+=1
    soup = getAndParseURL(user)
    
    if soup.find('div', {'class': 'profile_header_valve_employee'}):
        print(f'{count}: Valve Employee! User: {user}')
        continue
    
    if soup.find('div', {'class': 'profile_private_info'}):
        print(f"{count}: {user}'s profile is private!")
        continue
    
    userId = re.search('([^\/]+$)', user).group(0)
    users_data = {'userId': userId}

    profile_header = soup.find('div', {'class': 'header_real_name ellipsis'})
    if profile_header:
        userName = profile_header.bdi.extract().text
        if soup.find('span', {'class': 'actual_persona_name'}):
            userName = soup.find('span', {'class': 'actual_persona_name'}).text
        country = profile_header.text.strip()
        users_data |= {'User_Name': userName, 'Country': country}
    else:
        print(f'{count}: User: {user} has no profile header!')
    
    games_data=[]
    games_link = soup.find('a', {'href': user+'/games/?tab=all'})
    if 'Games' in soup.find('div', {'class': 'profile_item_links'}).text:
        soup_games = getAndParseURL(games_link['href'])
        var_rgGames = soup_games.find('script', {'language': 'javascript'})
        if var_rgGames:
            games = re.search('(?<=\[{)(.*)(?=}\])', var_rgGames.text)
            if games:
                games = games.group(0)
                games_list = re.split('},{', games)
                for game in games_list:
                    gem_gameId = re.search('(?<="appid":)(\d*)', game).group(0)
                    title = re.search('(?<="name":")(.*?)(?=")', game).group(0)
                    playTime = re.search('(?<="hours_forever":")(.*?)(?=")', game)
                    if playTime:
                        hoursPlayed = playTime.group(0)
                    else:
                        hoursPlayed = '0'

                    if float(hoursPlayed.replace(',', '')) > 0:
                        games_data.append({'userId': userId,
                                           'gameId': gem_gameId, 'Title': title, 'Hours_Played': hoursPlayed})
                if len(games_data) > 0:
                    user_games_collection.insert_many(games_data)
                else:
                    print(f'{count}: User: {user} has no games!')
        else:
            print(f'{count}: User: {user} has no var_rgGames javascript!')
    else:
        print(f'{count}: User: {user} has no games!')
    
    reviews_data=[]
    reviews_link = soup.find('a', {'href': user+'/recommended/'})
    if reviews_link:
        pages = [f'{reviews_link["href"]}?p=1']
        soup_reviews = getAndParseURL(pages[0])

        other_pages = soup_reviews.findAll('a', {'class': 'pagelink'})
        if other_pages:
            last_page = int(other_pages[-1]['href'][3:])
            for links in range(2, last_page+1):
                pages.append(user+"/recommended/?p="+str(links))
                
        for page in pages:
            soup_reviews = getAndParseURL(page)
            reviews_list = soup_reviews.findAll('div', {'class': 'review_box'})
            for reviews in reviews_list:
                rev_gameId = re.search('([^\/]+$)', reviews.find('div', {'class': 'leftcol'}).a['href']).group(0)
                recommendation = reviews.find('div', {'class': 'title'}).text
                reviewTime = re.findall('\d.+? ', reviews.find('div', {'class': 'hours'}).text.strip())[-1].strip()
                comment = reviews.find('div', {'class': 'content'}).text.strip()

                reviews_data.append({'userId': userId,
                                     'gameId': rev_gameId, 'Recomendation': recommendation,
                                     'Review_Time': reviewTime, 'Comment': comment})
        if len(reviews_data) > 0:
            user_reviews_collection.insert_many(reviews_data)
        else:
            print(f'{count}: User: {user} has no reviews!')
    else:
        print(f'{count}: User: {user} has no reviews!')

    user_collection.insert_one(users_data)

    if len(games_data) > 0 and len(reviews_data) > 0:
        print(f'{count}: {userId}...Done')
```
