```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.action_chains import ActionChains

import time

import re

import pandas as pd
import numpy as np

import pymongo

from dotenv import load_dotenv
import os
```

```python
driver = webdriver.Firefox(service=Service("../project_env/geckodriver"))
```

```python
url="https://store.steampowered.com/search/?filter=globaltopsellers&os=win"
driver.get(url)
```

```python
games_panel = driver.find_element(By.XPATH, '//div[@id="search_resultsRows"]')
games = games_panel.find_elements(By.TAG_NAME, 'a')
```

```python
count=0
game_count=0
price_data=[]
while game_count < 5000:
    time.sleep(.5)
    for game in games[count:]:
        count+=1
        if not(game.get_attribute('style') == "display: none;"):
            ActionChains(driver).move_to_element(game).perform()
            time.sleep(.5)
            gameId = re.search('(?<=\/)(\d+)', game.get_attribute('href')).group()
            print(f"Game exists: {gameId}")
            game_count+=1
            driver.execute_script('window.scrollBy(0, 52)')
            price = game.find_element(By.XPATH,
                    './/div[@class="col search_price_discount_combined responsive_secondrow"]')\
            .get_attribute('data-price-final')
            price = float(price[:-2]+'.'+price[-2:])
            price_data.append({'gameId': gameId, 'Price': price})
        else:
            print("Game does not exists!")
        if (count % 20 == 0):
            games = games_panel.find_elements(By.TAG_NAME, 'a')
```

```python
print(f'Count: {count}')
print(f'Game Count: {game_count}')
hidden_contents = driver.find_element(By.XPATH, '//div[@id="global_hover_content"]')
hidden_games = hidden_contents.find_elements(By.XPATH, './div')
print(f'Hidden Games: {len(hidden_games)}')
```

```python
info_data = []
count=0
for hidden_game in hidden_games:
    gameId = np.nan
    title = np.nan
    price = np.nan
    date = np.nan
    review = []
    tags = []
    
    count+=1
    
    gameId = re.search('\d+', hidden_game.get_attribute('id')).group()
    
    title_exists = hidden_game.find_element(By.TAG_NAME, 'h4')
    if title_exists:
        title = title_exists.get_attribute('textContent')
        print(f'{count}: {title}')
    
    date_exists = hidden_game.find_elements(By.XPATH, './div[@class="hover_release"]/span')
    if date_exists:
        date = date_exists[0].get_attribute('textContent')
    
    review_exists = hidden_game.find_elements(By.XPATH, './div[@class="hover_body"]')
    if review_exists:
        review = review_exists[0].get_attribute('textContent').strip()[25:].split('\n\t\t\t\t\t\t')[:2]
    
    for tag in hidden_game.find_elements(By.XPATH, './/div[@class="app_tag"]'):
        tags.append(tag.get_attribute('textContent'))

    data.append({'gameId': gameId, 'Title': title, 'Release Date': date, 'Review': review, 'Tags': tags})
```

```python
df = pd.merge(df_data, df_price, on='gameId')
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna('', inplace=True)
```

```python
data = df.to_dict('records')
```
